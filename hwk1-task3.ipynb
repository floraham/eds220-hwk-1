Brief dataset description: These are data on coral cover and number of “genera” in the Western Indian Ocean collected from a rapid assessment method used to sample coral bleaching.  The data includes the coral reef locations, number of coral taxa, and climate change, water quality, human resource management and extraction variables for each 6.25km2 reef cell averaged over the 1998-2022 sampling period. 

Study abstract: 
The study aimed to develop a spatial ecological model to predict the current and future numbers of coral species in the western Indian Ocean, considering climate change, water quality, human resource extraction, and variable selection processes. MLM were built based on 35 environmental factors using data from 7039 reef cells and ~1000 field surveys. Temperature variables were frequently important, but other factors like depth, wave energy, oxygen levels, and more also played crucial roles. Human influences were less influential compared to climate factors. Comparing climate scenarios, the study found that coral taxa losses were lower than coral cover losses over 30 years, and the effects of excess heat and temperature rise on corals were less severe than previously thought.
