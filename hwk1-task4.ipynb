# 7

```python
# solve exercise here
#a create a var top20 with 20 earthquakes
top20 = eqk.nlargest(20, "mag")
#b create bar plot showing the magnitude of the top 20 earthquakes. Update title, x-axis label, y-axis label, and color of the bars
top20.mag.plot.bar(title = "magnitude of top 20 earthquakes (by eqk ID)" , xlabel = "Earthquake ID" , ylabel = "Magnitude", color = "brown")

```


answer: 

The top20 index serves as an identifier for the earthquake. If we had not sorted through and updated the index we would have lost the information needed to identify the top 20 earthquakes in the original eqk dataset, sicne the index associated with the graph is not so meaningful in this context.
